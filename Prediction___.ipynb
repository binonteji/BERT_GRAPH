```python
import networkx as nx

stacked_G_train = nx.from_pandas_edgelist(unique_edges, 'node1', 'node2')
```

```python
from torch_geometric.utils import negative_sampling
from torch_geometric.utils.convert import from_networkx

edge_index_unified_train = from_networkx(stacked_G_train)
```

```python
def generate_data(emb, posEdges, negEdges):
    #stack codings of two proteins together
    # print('Hererrererererere')
    # print("Pos Edges : ", posEdges)
    # print(negEdges.shape)

    # print(posEdges)
    
    
    posNum = posEdges.shape[0]
    negNum = negEdges.shape[0]

    X = np.empty((posNum+negNum,2*emb.shape[1]))

    k = 0

    for x in posEdges:
#         print(emb[x[0]])
        X[k] = np.hstack((emb[x[0]],emb[x[1]]))
        k = k + 1
    for x in negEdges:
        X[k] = np.hstack((emb[x[0]],emb[x[1]]))
        k = k + 1 

    Y_pos = np.full((posNum,2),[0,1])
    Y_neg = np.full((negNum,2),[1,0])

    # print("Here........it is ", Y_pos.shape)

    Y = np.vstack((Y_pos,Y_neg))

    return X,Y



train_pos_Edges = edge_index_unified_train.edge_index.T
train_neg_Edges = negative_sampling(train_pos_Edges.T, edge_index_unified_train.num_nodes).T
```

```python
from keras.models import Sequential
from keras.layers import Dense,Activation,Dropout
import numpy as np

import warnings
warnings.filterwarnings("ignore")


def train_nn(X_train, Y_train, X_test, Y_test):
    mdl = Sequential()
    mdl.add(Dense(128, activation='relu', input_dim=X_train.shape[1]))
    mdl.add(Dropout(0.5))
    mdl.add(Dense(64, activation='relu'))
    mdl.add(Dropout(0.5))
    mdl.add(Dense(32, activation='relu'))
    mdl.add(Dropout(0.5))
    mdl.add(Dense(2,activation='softmax'))

    mdl.compile(loss='categorical_crossentropy',
                    optimizer='adam',
                    metrics=['accuracy'])
    mdl.fit(X_train, Y_train, epochs=50, batch_size=128, verbose=1)

#     mdl.save('trained_PPI_1767_mlp.h5')

    y_prob = mdl.predict(X_test)
    y_classes = y_prob.argmax(axis=-1)
    #print(y_classes)
    y_true = Y_test[:,1]
    print(y_true)

    return y_classes, y_true

pred, true  = train_nn(X_train,Y_train, X_test, Y_test)
```
